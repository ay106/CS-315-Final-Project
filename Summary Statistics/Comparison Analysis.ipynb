```python
import json
import os
import csv
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import re
```

```python
df1 = pd.read_csv('/Users/ryanboone/Desktop/CS_315/CS-315-Final-Project-main/compiling_data/final_data_by_year/final_result_2023.csv')
df2 = pd.read_csv('/Users/ryanboone/Desktop/CS_315/CS-315-Final-Project-main/compiling_data/final_data_by_year/final_result_may.csv')
```

```python
total_search_terms_df1 = len(df1['search_term'].unique())
total_search_terms_df2 = len(df2['search_term'].unique())
```

```python
common_search_terms = set(df1['search_term']).intersection(df2['search_term'])

# Calculate overlap percentage for each search term
overlap_percentages = {}
for term in common_search_terms:
    count_df1 = df1[df1['search_term'] == term]['search_term'].count()
    count_df2 = df2[df2['search_term'] == term]['search_term'].count()
    overlap_percentages[term] = (min(count_df1, count_df2) / max(count_df1, count_df2)) * 100
```

```python
plt.figure(figsize=(10, 8))
plt.barh(list(overlap_percentages.keys()), overlap_percentages.values(), color='skyblue')
plt.xlabel('Overlap Percentage')
plt.ylabel('Search Terms')
plt.title('Overlap Percentage of Search Terms between 2023 Files')
```

```python
df1 = pd.read_csv('/Users/ryanboone/Desktop/CS_315/CS-315-Final-Project-main/compiling_data/final_data_by_year/final_result_2023.csv')
df2 = pd.read_csv('/Users/ryanboone/Desktop/CS_315/CS-315-Final-Project-main/compiling_data/final_data_by_year/final_result_may.csv')
```

```python
total_search_terms_df1 = len(df1['search_term'].unique())
total_search_terms_df2 = len(df2['search_term'].unique())
```

```python
common_search_terms = set(df1['search_term']).intersection(df2['search_term'])

# Calculate overlap percentage for each search term
overlap_percentages = {}
for term in common_search_terms:
    count_df1 = df1[df1['search_term'] == term]['search_term'].count()
    count_df2 = df2[df2['search_term'] == term]['search_term'].count()
    overlap_percentages[term] = (min(count_df1, count_df2) / max(count_df1, count_df2)) * 100
```

```python
sorted_overlap_percentages = {k: v for k, v in sorted(overlap_percentages.items(), key=lambda item: item[1])}

# Plot the bar chart
plt.figure(figsize=(10, 8))
plt.barh(list(sorted_overlap_percentages.keys()), sorted_overlap_percentages.values(), color='skyblue')
plt.xlabel('Overlap Percentage')
plt.ylabel('Search Terms')
plt.title('Overlap Percentage of Search Results between 2023 Files')
```
