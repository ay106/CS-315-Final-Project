```python
import pandas as pd
import os
import re
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn.objects as so
```

## 1. Create dataframes of the csv's <b>
- Make sure to change the folder path 

```python
folder_path = r"C:\Users\ashle\OneDrive\School\CS315_Final\final_data"

dataframes = {} #dictionary will hold all the dataframes accessable by key

for filename in os.listdir(folder_path):
    if filename.endswith('.csv'):
        year_full = re.search(r'\d{4}', filename)
        year = year_full.group(0)
        file_path = os.path.join(folder_path, filename)
        dataframes[year] = pd.read_csv(file_path)

dataframes.keys()
```

## 2. Subjectivity Visualization <b>


```python
def summary(dataframes):
    for key in dataframes.keys():
        df = dataframes[key]
        none_count = df['Subjectivity/Objectivity'].isna().sum()
        print(f'NUmber of total videos in {key}: {len(df)}')
        print(f"Number of None values in 'Subjectivity/Objectivity': {none_count}\n")
summary(dataframes)
```

```python
table = {
    'year':[2020,2021,2022,2023,2024],
    'keywords': [["Quarantines","Coronavirus (2019-nCoV)","Disaster Relief"], ["Capitol Building (Washington, DC)","Storming of the US Capitol (Jan, 2021)","House Committee on Homeland Security"],[],["Israeli-Palestinian Conflict","Zionism","Palestinian Refugee"],["Presidential Election of 2024","Presidential Campaigns","Inaugural Address"]],
    'Total Videos':[2876,2492,0,2723,3070],
    'Subjective Videos': [2787,2411,0,2602,3001],
    'Objective Videos':[70,64,0,87,63],
    'Not Sure': [19,17,0,34,6],

    }
pd.set_option('max_colwidth', 200)
df = pd.DataFrame(table)
df.head()
```

```python
def keywords(dataframes):
    for key in dataframes.keys():
        df = dataframes[key]
        unique_names = df['search_term'].unique()
        print(f'Keywords in {key}: {unique_names}\n')
keywords(dataframes)

```

```python
df_2020= dataframes['2020']
sns.set_theme()
ax = sns.countplot(df_2020, x="Subjectivity/Objectivity", hue="author_verified")
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, ['Verified', 'Not Verified'], title='Author Status')
plt.title('Video Classification and User Verification Overview')
plt.xlabel('Transcription Classification')
plt.ylabel('Number of Videos')
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()), 
                 ha = 'center', va = 'center', xytext = (0, 9), textcoords = 'offset points')

```
