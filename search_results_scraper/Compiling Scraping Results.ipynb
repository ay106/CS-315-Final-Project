```python
import json
import os
import csv
import pandas as pd
```

```python
#opening csv results file and downloading as pandas 
df = pd.read_csv('test_results.csv')
df.head(2)
```

```python
#open json file and read in as dictionary
with open ('pyktok_test.json', 'r') as f:
    data = json.load(f)
data
```

```python
#get dict with search term and video ID as key and value
def get_ID_search_term(json_dict):
    result = {}
    for key in json_dict:
        # key is each search term
        all_urls = json_dict[key]['urls']
        for url in all_urls:
            #get video ID
            vid_ID = int(url[len(url)-19:])
            result[vid_ID] = key
    return result
```

```python
id_search_dct = get_ID_search_term(data)
```

```python
#iterate through df rows and add seach term to the csv
search_term_column = []
for index, row in df.iterrows():
    try:
        id = row['video_id']
        search_term_column.append(id_search_dct[id])
    except  Exception as e:
        print(f'{e}')
    
df['search_term'] = search_term_column
```
