```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df2020 = pd.read_csv('/Users/audreyliang/cs315/CS-315-Final-Project/compiling_data/final_data_by_year/final_result_2020.csv')
df2021 = pd.read_csv('/Users/audreyliang/cs315/CS-315-Final-Project/compiling_data/final_data_by_year/final_result_2021.csv')
df2022 = pd.read_csv('/Users/audreyliang/cs315/CS-315-Final-Project/compiling_data/final_data_by_year/final_result_2022.csv')
df2023 = pd.read_csv('/Users/audreyliang/cs315/CS-315-Final-Project/compiling_data/final_data_by_year/final_result_2023.csv')
df2024 = pd.read_csv('/Users/audreyliang/cs315/CS-315-Final-Project/compiling_data/final_data_by_year/final_result_2024.csv')
```

```python
search = df2020['search_term'].unique()
search.sort()

indices = {term: index for index, term in enumerate(search)}
matrix = np.zeros((len(search), 50), dtype=int)

for _, row in df2020.iterrows():
    search_index = indices[row['search_term']]
    
    video_index = row.name % 50  # 'name' here is the index of the row in the DataFrame
    matrix[search_index, video_index] = int(row['author_verified'])

# Create the heatmap
plt.figure(figsize=(20, 20))
sns.heatmap(matrix, cmap=['navy', 'red'], cbar_kws={'label': 'Verified Account'}, yticklabels=search)
plt.title('Heatmap of Verified Account Presence by Search Term and Video Index in 2020')
plt.xlabel('Video Index')
plt.ylabel('Search Term')
plt.xlim(0, 50)
plt.show()
```
